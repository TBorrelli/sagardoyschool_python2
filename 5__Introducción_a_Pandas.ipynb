<h2>Métodos de previsualización</h2>

El viaje de exploración de datos comienza con el recorrido básico de los datos - cómo se ven, qué columnas contienen, cuáles son los diferentes tipos de datos de las columnas, y así sucesivamente. <br>
Las funciones de previsualización de Pandas nos permiten echar un vistazo a los datos. <br>
El método <code>head()</code> muestra las primeras filas de datos (por defecto 5 pero se puede controlar). <br>
Este método toma un argumento opcional para el número de filas que el usuario quiere ver

```python
titanic_data.head() #por defecto 5 
```

```python
titanic_data.head(15) #ahora primeras 15 líneas 
```

El método <code>tail()</code> muestra las últimas filas de datos (por defecto 5) y de forma similar a <code>head()</code> toma como argumento el número de filas.<br>

```python
titanic_data.tail() #por defecto 5
```

```python
titanic_data.head(15) #ahora últimas 15 líneas 
```

Por último, el método <code>sample()</code> muestra las filas de datos muestreadas aleatoriamente (por defecto 1) del <code>dataset</code> (conjunto de datos). <br>
Por ejemplo, si desea comprobar 10 filas de muestra aleatoria del <code>dataset</code> (conjunto de datos) titanic, puede hacerlo de la siguiente manera:

```python
titanic_data.sample(10) #10 líneas aleatorias
```

<h2>Estadística descriptiva</h2>

Calcular estadísticos descriptivos como la <code>central tendency</code> (tendencia central), la <code>data dispersion</code> (dispersión de datos), el <code>percentil</code> y los valores <code>NaN</code> (Not a Number) ayuda a identificar si los datos son adecuados para un uso posterior. <br>
Calcularlos uno a uno usando diferentes métodos puede ser tedioso, por lo que Pandas proporciona el método <code>describe()</code> que puede calcular todos estos estadísticos:

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ La medida de tendencia central (moda, media y mediana), parámetro de una tendencia central o medida de centralización es un número ubicado hacia el centro de la distribución de los valores de una serie de observaciones (medidas), en la que se encuentra ubicado el conjunto de los datos<br>
Si quieres tener más información sobre la tendencia central visita <a href="https://es.wikipedia.org/wiki/Medidas_de_tendencia_central">este enlace</a> 
</div>

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ En estadística, las medidas de dispersión (también llamadas variabilidad, dispersión o propagación) es el grado en que una distribución se estira o se comprime<br>
Si quieres tener más información sobre las medidas de dispersión <a href="https://es.wikipedia.org/wiki/Medidas_de_dispersi%C3%B3n">este enlace</a> 
</div>

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ El percentil es una medida de posición usada en estadística que indica, una vez ordenados los datos de menor a mayor, el valor de la variable por debajo del cual se encuentra un porcentaje dado de observaciones en un grupo<br>
Si quieres tener más información sobre el percentil <a href="https://es.wikipedia.org/wiki/Percentil">este enlace</a> 
</div>

```python
titanic_data.describe()
```

<h2>Filtrado y Subconjunto de datos</h2>

<h3>Selección de columnas y filas específicas</h3>

A veces es posible que desee comprobar qué valores contiene una característica en el <code>dataset</code> (conjunto de datos), puede hacerlo utilizando el mecanismo de índice proporcionado por Pandas. <br>
Por ejemplo, digamos que queremos profundizar en los datos de <code>Age</code> (edad).<br>
Para acceder a la columna code>Age</code> (edad) de los datos de titanic, puede utilizar la siguiente línea de código:

Además, puede <code>slice</code> (trocear) esta columna para seleccionar el conjunto de valores utilizando la indexación tipo lista de la siguiente manera:

```python
titanic_data['Age'][:5]
```

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ Recuerde que ya hemos utilizado <code>slice()</code> tanto en las <code>string</code> como en la <code>list</code> y la <code>tuple</code><br>
</div>

Cuando se buscan las filas, es necesario trabajar sobre la columna índice del <code>dataset</code> (conjunto de datos). <br>
La columna índice por defecto para cualquier dataframe cargado con Pandas tiene valores de <b>1 a N (número de filas)</b> hasta que se establece explícitamente. <br>
Para acceder a los valores de las filas Pandas proporciona la función <code>loc[]</code> que toma el valor del índice como argumento.<br><br>


Para este ejemplo vamos a hacer que nuestra columna <code>Nombre</code> sea el índice para entender mejor el método <code>loc[]</code>

```python
# Establecer la columna Nombre como índice
titanic_data_index = titanic_data.set_index('Name')

# comprueba los valores con el valor del índice
titanic_data_index.loc['Wilkes, Mrs. James (Ellen Needs)']
```

Como puedes ver, se imprimen todos los valores cuyo índice es <code>Wilkes, Sra. James (Ellen Needs)</code><br>
También puede proporcionar un rango de valores (utilizando el operador <code>:</code>) y <code>los nombres de las columnas para filtrar</code> las columnas deseadas.

```python
#Quiero ver las siguientes columnas: Survived, Age de todos los registros desde 'Wilkes, Mrs. James (Ellen Needs) hasta Ware, Mr. Frederick
titanic_data_index.loc['Wilkes, Mrs. James (Ellen Needs)' : 'Ware, Mr. Frederick', ['Survived', 'Age']]
```

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ Si abres el <code>dataset</code> (conjunto de datos) (TitanicDataset.csv) verás que Wilkes, <code>Mrs. James (Ellen Needs)</code> es la línea <code>1</code> y <code>Ware, Mr. Frederick</code> es la línea <code>417</code>

<h3>Utilización del filtrado condicional</h3>

El filtrado condicional es una de las partes más importantes de la exploración de datos, ya que le ayuda a descubrir diferentes patrones en los datos basados en filtros. <br>
Para el filtrado condicional, dispone de diferentes operaciones de comparación como <code>==</code>, <code>></code>, <code><</code> y <code>!=</code> junto con operadores lógicos como <code>and</code>, <code>or</code> y <code>not</code>. <br>
La expresión básica de filtrado requiere seleccionar una <code>columna</code> y un valor de <code>filtro</code> o una lista de valores de filtro entre corchetes <code>[]</code>.<br>
Por ejemplo, filtrar la <code>Pclass</code> (clase de pasajero) en el <code>dataset</code> (conjunto de datos) titanic donde el valor de clase es <code>3</code>, tendrá la siguiente expresión de filtrado:

```python
titanic_data[titanic_data['Pclass'] == 3]
```

También puede aplicar varias condiciones a una sola columna o a un conjunto de columnas utilizando los operadores lógicos.<br>
Por ejemplo, filtrar los datos de titanic donde <code>Pclass</code> (clase de pasajero) es <code>3</code> y <code>Age</code> (edad de los pasajeros) es mayor o igual a <code>30</code>

```python
titanic_data[(titanic_data['Pclass'] == 3) & (titanic_data['Age'] >= 30)]
```

También puede aplicar la función Python preconstruida para el filtrado numérico y de <code>string</code>, como <code>isin()</code>, <code>between()</code>, <code>str.contains()</code>, <code>isna()</code>, <code>str.startswith()</code> y <code>str.endswith()</code>.<br>
Por ejemplo, la selección de las filas del <code>dataset</code> (conjunto de datos) original de titanic en las que el <code>Name</code> (nombre) empieza por <code>K</code>:

```python
titanic_data[titanic_data['Name'].str.startswith('K')]
```

<h3>Recuento de valores</h3>

Proporciona el recuento de <code>valores únicos</code> en una columna. <br>
Un ejemplo de cálculo de los recuentos de valores en la columna <code>Sex</code> (sexo) de los datos titanic se verá así:

```python
titanic_data['Sex'].value_counts()
```

También se pueden calcular los recuentos de valores multiíndice en el <code>dataset</code> (conjunto de datos). <br>
Por ejemplo, basándose en <code>Pclass</code>, la distribución de los valores de la característica <code>Age</code> (Edad) puede calcularse del siguiente modo:

```python
titanic_data[['Pclass', 'Age']].value_counts()
```

<h3>Agrupación y Agregación</h3>

A veces también puede necesitar agrupar los datos basándose en diferentes estadísticas como la <code>mean</code> (media), la <code>sum</code> (suma) y la <code>median</code> (mediana) que revelan muchos patrones ocultos en los datos. <br>
El método <code>groupby()</code> de Pandas ayuda a agrupar el conjunto más grande de datos y permite realizar diferentes operaciones en él. <br>
Para utilizar el método <code>groupby()</code>, debe proporcionar el <code>groupby()</code>nombre de la columna (o columnas) y una función agregada basada en la cual se agruparán los datos. <br>
Por ejemplo, si necesitamos agrupar los datos de titanic basándose en la columna <code>Sex</code> (Sexo) y con la función de agregación <code>mean()</code>  (media) se puede hacer en la siguiente manera:

```python
titanic_data.groupby('Sex').mean(numeric_only=True)
```

Con <code>groupby</code>, podemos ver que el valor medio de mujeres que sobrevivieron al accidente del Titanic es <code>1</code><br>
Esto es cierto, ya que la mayoría de las mujeres a bordo sobrevivieron al incidente (aunque muestra las limitaciones de la media como estadística central).<br>
La agrupación de datos también puede ser multiíndice, en la que se pueden agrupar datos basándose en dos o más categorías

```python
titanic_data.groupby(['Sex', 'Pclass']).mean(numeric_only=True)
```

Pandas también ofrece la posibilidad de crear gráficos y diagramas directamente con el <code>dataset</code> (conjunto de datos).<br>
Por ejemplo, crear un gráfico de barras con los datos agregados, se puede mostrar llamando al método <code>plot()</code> de la siguiente manera:

```python
titanic_data.groupby(['Sex']).mean(numeric_only=True).plot(kind='bar')
```

Veremos más sobre la visualización de datos en el próximo capitulo 

El agrupamiento condicional también es posible usando la función <code>loc[]</code> en Pandas. <br>
En la función <code>loc[]</code>, primero proporcionamos la <code>condición de filtrado</code> y luego las <code>condiciones groupby</code>:

<img src="data:image/png;base64, "  align="center" />

```python
titanic_data.loc[titanic_data['Age']>25, ['Pclass', 'Survived']].groupby(['Pclass', 'Survived'])['Survived'].count()
```

Es posible agregar funciones como <code>mean()</code>, <code>sum()</code>, <code>count()</code> en diferentes columnas del <code>dataset</code> (conjunto de datos) utilizando el método <code>aggregate()</code> de Pandas.

```python
titanic_data.groupby(['Pclass', 'Survived'] )['Survived'].agg(["count", "mean"])
```

<h2>Limpieza y preprocesamiento de datos</h2>

Ahora vamos con la parte más importante del análisis exploratorio de datos que ayuda a dar forma a tus datos para el análisis ML (Machine Learning).<br>
Los datos del mundo real no siempre son ideales y no pueden utilizarse directamente para el análisis ML (Machine Learning), ya que pueden contener valores <code>NaN</code> (Not a Number), datos inexactos, valores duplicados o valores corruptos. <br>

<h3>Identificación de <code>Missing Value</code> (valores perdidos) y tratamiento de los mismos</h3>

La mayoría de los modelos de <code>ML</code> (aprendizaje automático) y <code>deep learning</code> (aprendizaje profundo) no pueden manejar valores <code>NaN</code>, por lo que es necesario manejarlos antes de pasar los datos a estos modelos. <br>
Tres enfoques normales para manejar estos valores NaN son:
<ul>
    <li><b>Predicción</b> de los valores <code>NaN</code> basada en las demás características del <code>dataset</code> (conjunto de datos).</li>
    <li><b>Eliminar</b> los valores <code>NaN</code> del <code>dataset</code> (conjunto de datos) ⚠️ <b>no se recomienda a menudo</b>.</li>
    <li><b>Reemplazar</b> los valores <code>NaN</code> por los valores más adecuados (a menudo con la medida de tendencia central).</li>
</ul>

Pandas proporciona el método <code>dropna()</code> para eliminar valores nulos basados en filas y columnas. <br>

 Te habrás dado cuenta de que las columnas <code>Age</code> (Edad), <code>Fare</code> (Tarifa) y <code>Cabin</code> (Cabina) contienen valores <code>NaN</code>titanic_data_dropna = titanic_data.dropna(axis=1)titanic_data_dropna.info()<br>
 Para eliminar estos valores puede utilizar el siguiente código

```python
titanic_data_dropna = titanic_data.dropna(axis=1)
titanic_data_dropna.info()
```

<div class="alert alert-success alertsuccess" style="color:black">
El argumento <code>axis</code> controla las características de las filas y columnas. <br>
    Si se establece en <code>0</code>, se elimina toda la fila de los datos. <br>
    Si se establece en <code>1</code>, se elimina toda la columna que contenga cualquier valor <code>NaN</code>
</div>

Otro método proporcionado por Pandas es <code>fillna()</code> que reemplaza los valores <code>NaN</code> por los valores proporcionados.<br>
Por ejemplo, si desea reemplazar los valores <code>NaN</code> en la columna <code>Age</code> (Edad) por una edad media en el conjunto de datos, puede hacerlo de la siguiente manera:

```python
# crear una copia del conjunto de datos para probar el método fillna
titanic_data_fillna = titanic_data.copy()

# sustituye los valores NaN por la edad media
titanic_data_fillna.fillna(value=titanic_data_fillna['Age'].mean(),inplace=True)

# comprueba los valores NaN sustituidos
titanic_data_fillna.tail()
```

<div class="alert alert-block alert-info" style="color:black">
<b>Nota Bene:</b> El argumento <code>inplace</code> permite realizar cambios en la misma instancia del marco de datos
</div>

<h3>Unir diferentes conjuntos de datos</h3>

Unir diferentes conjuntos de datos para diferentes casos de uso es una necesidad muy común. <br>
Puede utilizar el método <code>concat()</code> de Pandas para unir diferentes conjuntos de datos basados en filas y columnas. <br>
Esta función sólo requiere una lista de dataframes y el parámetro <code>axis</code> que determina la orientación de la fusión, es decir:<br>
<ul>
    <li>si 0 dataframes se fusionarán en filas</li>
    <li>si 1 dataframes se fusionarán en columnas</li>
</ul>
Por ejemplo, concatemos el conjunto de datos titanic consigo mismo:

```python
duplicate_data = pd.concat([titanic_data, titanic_data], axis=0)
duplicate_data.shape
```

Unir los dataframes de diferentes dimensiones con el método <code>concat()</code> puede introducir valores <code>NaN</code> innecesarios.<br>
Para solucionar este problema Pandas también proporciona un método <code>join()</code> que funciona exactamente igual que las uniones <code>SQL</code><br>
El método <code>join()</code> requiere un parámetro <code>how</code> que determina el tipo de <code>join</code>, como <code>inner</code> o <code>outer</code>, y los parámetros on indican en qué columna se basará el <code>join</code>. <br>
Un ejemplo sencillo de unión de dos marcos de datos de ejemplo puede tener este aspecto:

```python
# fusionar dos conjuntos de datos
df_1 = pd.DataFrame({
	"columna_a": ["A", "B", "C", "D", "E"],
	"columna_b": [10, 11, 12, 13, 14]},
	index=["1", "2", "3", "4", "5"]
)
df_1.head()
```

```python
df_2 = pd.DataFrame({
	"columna_a": ["A", "B", "E"],
	"columna_c": [100, 101, 104]},
	index=["1", "2", "3"]
)
df_2.head()
```

```python
#inner join basados en la columna
pd.merge(df_1, df_2, on='columna_a', how='inner')
```
