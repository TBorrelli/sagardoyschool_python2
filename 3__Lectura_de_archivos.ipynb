<h1>Lectura de archivos de texto</h1>

Una forma de leer o escribir un archivo en Python es utilizar la función built-in (incorporado) <code>open</code><br>
La función <code>open</code> proporciona un **File Object** (objeto Archivo) que contiene los métodos y atributos necesarios para leer, guardar y manipular el archivo. <br>
Los parámetros que necesitas son el <code>File Path</code> (ruta), el <code>File Name</code> (nombre del archivo), el <code>Mode</code> (modalidad) y el <code>Encoding</code> (codificación). <br>

<h2 id="mejor">Una forma mejor de abrir un archivo</h2>

Utilizar la declaración <code>with</code> es una mejor práctica, cierra automáticamente el archivo incluso si el código encuentra una excepción. <br>
El código ejecutará todo en el bloque de indentación y luego cerrará el objeto de archivo.

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    # código a ejecutar con el fichero abierto
    # ...
    contenido_file = file1.read()
    # ...
    # código a ejecutar con el fichero abierto

# aquí, automáticamente el archivo se cierra
print(contenido_file)
```

El objeto archivo está cerrado, puede verificarlo ejecutando la siguiente celda:

```python
file1.closed
```

No tenemos por qué leer todo el archivo, por ejemplo, podemos leer los 4 primeros caracteres introduciendo 4 como parámetro del método **.read()**:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    print(file1.read(4))
```

Una vez que el método <code>.read(4)</code> lee los 4 primeros caracteres. <br>
Si volvemos a llamar al método <code>.read(4)</code> se leerá a los 4 caracteres siguientes.<br>
En la siguiente celda se demostrará el proceso para diferentes entradas al método <code>read()</code>:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    print("primera lectura: '"+file1.read(4)+"'")
    print("segunda lectura: '"+file1.read(4)+"'")
    print("tercera lectura: '"+file1.read(7)+"'")
    print("cuarta lectura: '"+file1.read(15)+"'")
```

El proceso se ilustra en la siguiente figura, y cada color representa la parte del archivo leída tras el método <code>read()</code>:

<img src="data:image/png;base64, " align="center" />"

Aquí hay un ejemplo usando el mismo archivo, pero en su lugar leemos 16, 5, y luego 9 caracteres a la vez:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    print("primera lectura: '"+file1.read(16)+"'")
    print("segunda lectura: '"+file1.read(5)+"'")
    print("tercera lectura: '"+file1.read(9)+"'")
```

También podemos leer una **entera** línea del archivo cada vez mediante el método <code>readline()</code>:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    print("primera línea: " + file1.readline())
```

También podemos pasar un argumento a <code> readline() </code> para especificar el número de caracteres que queremos leer. <br>
Sin embargo, a diferencia de <code> read()</code>, <code> readline()</code> sólo puede leer una línea como máximo.

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    print(file1.readline(20)) # no lee más allá del final de línea.
    print(file1.read(20)) # Devuelve los siguientes 20 caracteres.
```

Podemos utilizar un bucle para recorrer cada línea:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
        i = 0;
        for línea in file1:
            print("Iteración", str(i), ": ", línea)
            i = i + 1
```

Podemos utilizar el método <code>readlines()</code> para guardar el archivo de texto en una lista:

```python
with open(example1, mode='r',encoding='utf-8') as file1:
    file_como_lista = file1.readlines()
```

Cada elemento de la lista corresponde a una línea de texto:
