# Visualización de datos con Matplotlib

Matplotlib es una biblioteca de visualización muy básica pero rica en características que le proporciona opciones para crear diferentes gráficos y diagramas, incluyendo gráficos de líneas, gráficos de distribución, gráficos de barras, gráficos de dispersión y visualización avanzada como mapas geográficos de calor<br>
Es la biblioteca de visualización en Python viene de Matplotlib.

## Creación de un Line plot (gráfico lineal)

Para crear un gráfico lineal simple, puedes utilizar el método <code>plot()</code> de matplotlib.<br>
Puedes crear un gráfico con los valores del <code>eje X</code> y los valores del <code>eje Y</code>, o sólo con el <code>eje X</code>.<br>
Este gráfico se utiliza mucho para establecer una <u>relación entre dos o más variables</u>. <br>


Creamos un gráfico lineal simple del precios de las primeras 100 viviendas

```python
plt.plot(df['price'][:100])
```

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ Como puede ver, el eje de precios (Y) se expresa de forma científica. <br>
Este es el comportamiento por defecto de matplotlib. Se puede desactivar con la función <code>ticklabel_format()</code> y lo <code>style = 'plain'</code>
</div>

```python
plt.ticklabel_format(style='plain')
plt.plot(df['price'][:100])
```

Se puede personalizar aún más este gráfico con <code>markers</code> (marcadores), <code>colors</code> (colores), <code>xlabels</code> (etiquetas_x), <code>ylabels</code> (etiquetas_y), <code>titles</code> (títulos), etc., ya que matplotlib admite todas estas funcionalidades

```python
plt.plot(df['price'][:100], marker = 'o', color = 'green', label = 'Precios de 100 viviendas')
plt.title('Precios de 100 viviendas')
plt.xlabel('Precio Vivienda')
plt.legend()
```

## Creación de un Scatter plot (gráfico de dispersión)

Puede crear un gráfico de dispersión utilizando el método <code>scatter()</code> de matplotlib. <br>
Esto traza diferentes puntos de dispersión entre el <code>eje Y</code> y el <code>eje X</code>. <br>
Estos gráficos identifican la <u>relación, correlación y tendencia entre dos variables</u>

Un simple gráfico de dispersión entre las características de las plantas y el precio de los datos de una casa

```python
plt.scatter(df['price'], df['floors'], marker = 'o', color = 'green', label = 'Precio de la casa por plantas')
plt.title('Número de plantas vs Precio')
plt.xlabel('Precio vivienda')
plt.ylabel('Número de plantas')
plt.ticklabel_format(style='plain')
plt.legend()
```

## Creación de un Bar plot (gráfico de barras)

Los <code>Line plots</code> (gráficos de lineal) y <code>Scatter plots</code> (de dispersión) se limitan a los datos numéricos, por lo que si desea visualizar datos categóricos, los <code>Bar plot</code> (gráficos de barras) son la solución. <br>
Puedes utilizar el método <code>bar()</code> de matplotlib para representar los datos categóricos

Vamos a trazar un gráfico de barras para los precios de venta de viviendas por ciudad

```python
# Ajustar el tamaño de la figura
plt.figure(figsize=(10,5))
plt.bar(df['city'][:10],df['price'][:10], color='cyan')
# Crear un gráfico de barras con color cian


# Añadir etiquetas y título
plt.xlabel('Ciudad')
plt.ylabel('Precio de venta')
plt.title('Precio de venta por ciudad')


# Visualizar el gráfico
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ La primera línea del código, <code>figsize()</code> permite controlar el <code>canvas</code> (lienzo) donde se representa el gráfico, se puede ajustar el tamaño del <code>canvas</code> (lienzo) en función de las necesidades
</div>

```python
# Ajustar el tamaño de la figura
plt.figure(figsize=(20,3))
plt.bar(df['city'][:10],df['price'][:10], color='cyan')
# Crear un gráfico de barras con color cian


# Añadir etiquetas y título
plt.xlabel('Ciudad')
plt.ylabel('Precio de venta')
plt.title('Precio de venta por ciudad')


# Visualizar el gráfico
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

## Creación de un Histogram (gráfico de histograma)

Un gráfico similar que representa lo cuantitativo es un histograma. <br>
Este tipo de gráfico ayuda a identificar la distribución de frecuencias en los datos y se puede trazar con el método <code>hist()</code> 

```python
# Crear un histograma con color rojo
plt.hist(df['price'][:1000], color='red', bins = 100)


# Añadir etiquetas y título
plt.ylabel('Precio de venta')
plt.title('Distribución del precio de venta')


# Visualizar el gráfico
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ El argumento <code>bin</code> en el método <code>hist()</code> representa los intervalos en el gráfico, cuanto más altos son los bins más bajos son los intervalos 
</div>

```python
# Crear un histograma con color rojo
plt.hist(df['price'][:1000], color='red', bins = 10)


# Añadir etiquetas y título
plt.ylabel('Precio de venta')
plt.title('Distribución del precio de venta')


# Visualizar el gráfico
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

## Creación de subplots (cuadrícula de gráficos)

Si está creando visualizaciones para análisis o informes, a veces necesitará crear múltiples gráficos juntos para visualizar y comparar diferentes categorías.<br>
Subplots ayuda a definir una cuadrícula de gráficos (en filas y columnas) y permite crear gráficos en estas cuadrículas. <br>
Para ello es necesario utilizar el método <code>subplot()</code> para crear la <code>grid of plots</code> (cuadrícula de gráficos) (m X n)

```python
# Crear una figura con múltiples subparcelas
fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # 2 filas, 1 columna

# Primera subgrupo (izquierda)
axs[0].scatter(df['bedrooms'],df['price'], color='blue')
axs[0].set_title('Precio por número de dormitorios')
axs[0].set_xlabel('Número de habitaciones')
axs[0].set_ylabel('Precio')

# Segundo subgrupo (derecha)
axs[1].scatter(df['bathrooms'],df['price'], color='red')
axs[1].set_title('Precio por número de baños')
axs[1].set_xlabel('Número de baños')
axs[1].set_ylabel('Precio')

# Ajustar el espaciado entre los subgrupos
plt.tight_layout(pad=6)

# Mostrar los gráficos
axs[0].ticklabel_format(style='plain', axis='y')
axs[1].ticklabel_format(style='plain', axis='y')
plt.show()
```

## Creación de gráficos 3D

La creación de gráficos 3D y gráficos de superficie con Plotly es una forma eficaz de visualizar datos en tres dimensiones. <br>
Estos tipos de gráficos son particularmente útiles cuando se trata de datos que tienen múltiples variables y desea explorar las relaciones. <br>
He aquí un ejemplo de cómo crear gráficos 3D utilizando Plotly en Python

```python
import plotly.graph_objects as go

# Crear un gráfico de dispersión 3D
x = np.random.rand(100)
y = np.random.rand(100)
z = x**2 + y**2 # Generar un valor z basado en x e y

scatter_plot = go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=6))

layout = go.Layout(title='Diagrama de dispersión 3D')
fig_scatter = go.Figure(data=[scatter_plot], layout=layout)


# Mostrar los gráficos
fig_scatter.show()
```

```python
import plotly.graph_objects as go


# Crear un gráfico de dispersión 3D
x = np.random.rand(100)
y = np.random.rand(100)
z = x**2 + y**2  # Generar un valor z basado en x e y

# Crear un gráfico de superficie
x_surface = np.linspace(-5, 5, 50)
y_surface = np.linspace(-5, 5, 50)
x_surface, y_surface = np.meshgrid(x_surface, y_surface)
z_surface = np.sin(np.sqrt(x_surface**2 + y_surface**2))

surface_plot = go.Surface(x=x_surface, y=y_surface, z=z_surface)

layout_surface = go.Layout(title='Gráfico de superficie')
fig_surface = go.Figure(data=[surface_plot], layout=layout_surface)

# Mostrar los gráficos
fig_surface.show()
```

## Creación de Distribution Plots (gráficos de distribución)

También puede personalizar diferentes gráficos de distribución como KDE Plots o histogramas con Seaborn.<br> 
Por ejemplo, puede crear tanto un <code>histogram</code> (histograma) como un <code>KDE plot</code> (gráfico KDE) en la misma figura llamando a los métodos <code>histplot()</code> y <code>kdeplot()</code>

<div class="alert alert-success alertsuccess" style="color:black">
ℹ️ <b>Qué es una gráficos KDE?</b> El gráfico KDE se utiliza para estimar la función de densidad de probabilidad de una variable continua, proporcionando información sobre la distribución, la forma y la tendencia central de los datos<br>
Para más información, visitas <a href="https://www.geeksforgeeks.org/seaborn-kdeplot-a-comprehensive-guide/">este enlace</a>
</div>

```python
# Definir estilo Seaborn 
sns.set(style='whitegrid')


# Create a histogram (distplot)
plt.figure(figsize=(8, 6))
sns.histplot(df['price'][:1000], bins=20, color='blue')
plt.title('Gráfico de distribución (histograma)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
```

```python
# Definir estilo Seaborn 
sns.set(style='whitegrid')


# Crear un gráfico de estimación de densidad de núcleo (KDE)
plt.figure(figsize=(8, 6))
sns.kdeplot(df['price'][:1000], color='red')
plt.title('Gráfico de distribución (KDE)')
plt.xlabel('Valor')
plt.ylabel('Densidad')
```

Ahora vamos a crear un histograma y un gráfico KDE en la misma figura

```python
# Establecer estilo Seaborn
sns.set(style='whitegrid')

# Crear un gráfico combinado con histograma y KDE
plt.figure(figsize=(8, 6))
sns.histplot(df['price'][:1000], bins=20, kde=True, color='blue')
sns.kdeplot(df['price'][:1000], color='red')
plt.title('Gráfico de distribución (Histograma + KDE)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia / Densidad')
plt.show()
```

## Creación de Pair Plots y Heatmap (gráficos de pares y mapas de calor)

Con un <code>Pair Plots</code> (gráfico de pares), cada variable numérica de los datos se comparte a través de los ejes y en una sola fila y los ejes x en una sola columna. <br>
Esto le ayuda a comprobar la relación entre pares en los datos <br>
Para crear un <code>Pair Plots</code> (gráfico de pares), utiliza el método <code>pairplot()</code>y pása el dataframe para que genere un gráfico para todas las columnas numéricas

```python
# Establecer estilo Seaborn
sns.set(style='whitegrid')

# Crear un gráfico de pares
sns.pairplot(df)
plt.suptitle('Pair Plot', y=1.02)  # Añade un título encima del gráfico
plt.show()
```

Otra gran opción para explorar la relación entre las diferentes columnas del conjunto de datos es un <code>heat map</code> mapa de calor.<br>
Este gráfico muestra la correlación entre las distintas variables del conjunto de datos y puede representarse utilizando el método <code>heatmap()</code>

```python
# Calcular la matriz de correlaciones
correlation_matrix = df.corr(numeric_only=True)

# Crear un mapa de calor de la matriz de correlaciones
plt.figure(figsize=(12, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de calor de correlación')
plt.show()
```
