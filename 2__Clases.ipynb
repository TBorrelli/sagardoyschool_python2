<h1>Clases y Objetos</h1>

<h2>Creación de una clase</h2>

El primer paso para crear una clase es darle un nombre. <br>
Ahora, crearemos dos clases: <b>Círculo</b> y <b>Rectángulo</b>. <br>
Necesitamos determinar todos los datos que componen esa clase, a los que llamamos <em>attributes</em> (atributos).<br>
Piensa en este paso como la creación de un plano que utilizaremos para crear objetos. <br>
En la figura 1 vemos dos clases, <b>Círculo</b> y <b>Rectángulo</b>.
Cada una tiene sus atributos, que son variables.<br>
<ul>
    <li>La clase <b>Círculo</b> tiene el atributo radio y color</li>
    <li>la clase <b>Rectángulo</b> tiene el atributo altura, anchura y color</li>
</ul>
Vamos a utilizar los ejemplos visuales de estas formas antes de llegar al código, ya que esto le ayudará a acostumbrarse al vocabulario.

<h3>Métodos</h3>

Los métodos te dan una forma de cambiar o interactuar con el objeto; son funciones que interactúan con los objetos. <br>
Por ejemplo, digamos que nos gustaría aumentar el <code>radio</code> de un círculo en una cantidad especificada. <br>
Podemos crear un método llamado **cambia_radio(r)** que cambia el valor del <code>radio</code> en **r**. <br>
Esto se muestra en la figura, donde después de aplicar el método al "objeto círculo verde", el <code>radio</code> del objeto cambia en consecuencia. <br>

El primer paso para crear la clase es utilizar la palabra clave <code>class</code>, luego el nombre de la clase <br><br>

El siguiente paso es un método especial llamado constructor <code>\_\_init\_\_</code>, que se utiliza para inicializar el objeto. <br>
<code>self</code> contiene todos los atributos del objecto <br>
Por ejemplo el <code>self.color</code> te da el valor del atributo color y <code>self.radio</code> te dará el radio del objeto <br><br>

El método <code>cambia_radio</code> con el parámetro <code>r</code>, el cambia el radio del círculo al valor <code>r</code>. <br>
Para acceder al radio utilizamos la sintaxis <code>self.radio</code><br><br>

El método <code>dibujar_círculo</code> con el parámetro <code>self</code>, dibuja un círculo utilizando la librería <code>matplotlib</code>. <br>
La función de matplotlib: <code>gca()</code> crear una instancia 2D de un <code>canvas</code> (lienzo) vacío<br>
La función de matplotlib: <code>add_patch</code> dibuja un polígono en el lienzo y tiene como input: la función de matplotlib: <code>Circle</code> <br> 
la función <code>Circle</code> dibuja el círculo en los <code>puntos (0,0)</code> con <code>radius</code> = <code>self.radio</code> y <code>color</code> = <code>self.color</code><br>
La función de matplotlib: <code>axis('scaled')</code> ajusta el  <code>canvas</code> (lienzo) al tamaño del círculo<br>
La función de matplotlib: <code>show()</code> dibuja el resultado

<div class="alert alert-block alert-success" style="color:black">
<b>Nota Bene:</b> Por convención, el nombre de una clase empieza por mayúscula
</div>

<div class="alert alert-block alert-info" style="color:black">
<b>Nota Bene:</b> Fijamos el <code>radio</code> por defecto en <code>3</code> y el <code>color</code> por defecto en <code>blue</code> (azul)
</div>

<div class="alert alert-block alert-warning" style="color:black">
<b>Importante:</b> Como se dijo al principio del primer curso, deberíamos empezar a utilizar el inglés en lugar del español para la programación.<br>
Por ejemplo en este caso <code>color='blue'</code>, <code>plt.Circle</code>, <code>fc</code> (face color)... no se pueden traducir en español. <br>
La función solo aceptará términos en inglés.
</div>

````python
class Círculo():
    
    # Constructor
    def __init__(self, radio=3, color='blue'):
        self.radio = radio
        self.color = color 
        
    # Método cambia_radio
    def cambia_radio(self, r):
        self.radio = r
        return(self.radio)
    
    # Método dibujar_círculo
    def dibujar_círculo(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radio, fc=self.color))
        plt.axis('scaled')
        plt.show()  
````

<h2>La clase rectángulo</h2>

Vamos a crear una clase rectángulo con los atributos de <code>altura</code>, <code>anchura</code> y <code>color</code>. <br>
Sólo añadiremos el método para dibujar el objeto rectángulo <code>dibujar_rectángulo</code>:

````python
class Rectángulo(object):
    
    # Constructor
    def __init__(self, altura=2, anchura=3, color='r'):
        self.altura = altura 
        self.anchura = anchura
        self.color = color
    
    # Método dibujar_rectángulo
    def dibujar_rectángulo(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.anchura, self.altura ,fc=self.color))
        plt.axis('scaled')
        plt.show()
        
````

<b>Nota Bene:</b> Para establecer el <code>color</code> rojo utilizamos la abreviatura <code>r</code>
</div>

````python
class Coche:
    
    color = "blanco"
    def __init__(self, velocidad_máxima, kilometraje):
        self.velocidad_máxima = velocidad_máxima
        self.kilometraje = kilometraje
        self.capacidad_de_asientos = None

    def assign_seating_capacity(self, capacidad_de_asientos):
        self.capacidad_de_asientos = capacidad_de_asientos

    def mostrar_propiedades(self):
        print("Propiedades del coche:")
        print("Color:", self.color)
        print("Velocidad Máxima:", self.velocidad_máxima)
        print("Kilometraje:", self.kilometraje)
        print("Capacidad De Asientos:", self.capacidad_de_asientos)

Coche1 = Coche(200, 50000)
Coche1.assign_seating_capacity(5)
Coche1.mostrar_propiedades()

Coche2 = Coche(180, 75000)
Coche2.assign_seating_capacity(4)
Coche2.mostrar_propiedades()
````
