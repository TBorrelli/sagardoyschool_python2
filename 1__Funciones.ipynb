<h1>Funciones</h1>

Una función es un bloque de código reutilizable que realiza operaciones especificadas en la función. <br>
Permiten dividir las tareas y reutilizar el código en distintos programas.<br>


Existen dos tipos de funciones :


* <b>Funciones predefinidas</b>
* <b>Funciones definidas por el usuario</b>



<h2>¿Qué es una función?</h2>

Puede definir funciones para proporcionar la funcionalidad requerida. <br>
Aquí hay reglas simples para definir una función en Python: <br>

* Los bloques de funciones comienzan con <code>def</code> seguido de la función <code>name</code> y paréntesis <code>()</code>.
* Hay parámetros de entrada o argumentos que deben colocarse dentro de estos paréntesis.
* También puede definir parámetros dentro de estos paréntesis.
* Hay un cuerpo dentro de cada función que comienza con dos puntos (<code>:</code>) y tiene indentado.
* También se puede colocar documentación ante el cuerpo de la función.
* La declaración <code>return</code> sale de una función y, opcionalmente, devuelve un valor.

<h2>Ejemplo función</h2>

Un ejemplo de una función que se suma dos valores: <code>valore_a</code> y <code>valore_b</code> y imprime el resultado:

````python
def suma(valore_a, valore_b):
    
    """
    función que suma el parámetro valore_a al parámetro valore_b 
    """
    
    resultado = valore_a + valore_b
    print(valore_a, "+", valore_b, " = ", resultado)
    return(resultado)


suma(1,2)
````

Podemos crear diferentes funciones. <br>
Por ejemplo, podemos crear una función que multiplique dos números. <br>
Los números estarán representados por las variables <code>valore_a</code> y <code>valore_b</code>:

````python
def multiplication (valore_a, valore_b):
    resultado = valore_a * valore_b
    return(resultado)
    print('Este método print será ignorado porque está después de la declaración return')
    
resultado_multiplication = multiplication(12,2)
print(resultado_multiplication)
````

La misma función puede utilizarse para distintos tipos de datos. Por ejemplo, podemos multiplicar dos números enteros:

````python
multiplication(2, 3)
````

````python
multiplication(10.0, 3.14)
````

Incluso podemos replicar una string multiplicándola por un número entero:

````python
multiplication(2, "Tiziano ")
````

Podemos sustituir las líneas de código por una función. <br>

Una vez definida una función, puede utilizarse repetidamente. <br>

<h2>Usar <code>if</code>/<code>else</code> y bucles en funciones</h2>

La función <code>return()</code> es particularmente útil si tienes alguna declaración <code>if</code> en la función, cuando quieres que tu salida dependa de alguna condición:

````python
def nombre_y_rol(nombre, es_alumno):
    if es_alumno == True:
        return "Profesor "+ nombre
    else:
        return "Alumno" + nombre
x = nombre_y_rol("Tiziano", True)
print(x)
````

Podemos utilizar un bucle en una función. <br>
Por ejemplo, podemos <code>print()</code> cada elemento de una lista:

````python
def imprimirLista(lista):
    for elemento in lista:
        print(elemento)
````

````python
imprimirLista(['1', 1, 'Hola', "Felipe", True])
````

**Cuenta la frecuencia de las palabras que aparecen en una <code>string</code> utilizando un diccionario.**.

Supongamos que tenemos una <code>string</code> y <code>palabras</code> y tenemos que encontrar el recuento de ocurrencia de esta <code>testo_da_buscar</code> en la string utilizando python. <br>
Esto es lo que vamos a hacer en esta sección, contar el número de <code>testo_da_buscar</code> en una <code>string</code> dada e imprimirlo. <br>
1. Lo primero que haremos será definir una función
2. Crear una lista que estará vacía inicialmente.
3. Convertir la <code>string</code> en una lista. Python string tiene un método <code>split()</code>. Toma una <code>string</code> y algún separador para devolver una lista.
4. Diccionario vacío.
5. Bucle <code>for</code> para iterar las palabras y el valor contará la frecuencia de cada palabra en la <code>string</code> y las almacenará en el diccionario.
6. Print del diccionario.

````python
#1 Lo primero que haremos será definir una función
def conta_palabras(texto):
    
    #2 Crear una lista que estará vacía inicialmente.
    palabras = []
    
    #3 Convertir la string en una lista.
    palabras = texto.split() # or string.lower().split()
    
    #4 Diccionario vacío.
    diccionario = {}
    
    #5 Bucle <code>for</code> para iterar las palabras
    for clave in palabras:
        diccionario[clave] = palabras.count(clave)
        
    #6 Print del diccionario.
    print("La frecuencia de las palabras es:",diccionario)
    
# Test
conta_palabras("Con mi burrito sabanero, voy camino de Belén " \
               "Con mi burrito sabanero, voy camino de Belén " \
               "Si me ven, si me ven " \
               "Voy camino de Belén ")
````

<h2><code>*args</code> y <code>**kwargs</code></h2>

<h3><code>*args</code></h3>

Gracias a los <code>*args</code> en Python, podemos definir funciones cuyo número de argumentos es variable. <br>
Es decir, podemos definir funciones genéricas que no aceptan un número determinado de parámetros, sino que se “adaptan” al número de argumentos con los que son llamados. <br>
De hecho, el <code>*args</code> viene de arguments en Inglés, o argumentos. <br>
Haciendo uso de <code>*args</code> en la declaración de la función podemos hacer que el número de parámetros que acepte sea variable.

````python
def imprimir_todos_parametro(*args): # Todos los argumentos se 'empaquetan' en args que puede tratarse como una tupla.
    print("Nº de parámetros:", len(args))
    for parámetro in args:
        print(parámetro)

#printAll con 3 argumentos
imprimir_todos_parametro('Jose','Maria','Sandra')
#printAll con 4 argumentos
imprimir_todos_parametro('Jose','Tiziano','Maria','Sandra')
````

Antes de nada, el uso del nombre <code>*args</code> es totalmente arbitrario, por lo que podrías haberlo llamado como quisieras. <br>
Es una mera convención entre los usuarios de Python y resulta frecuente darle ese nombre. <br>
<b>Lo que si es un requisito, es usar * junto al nombre.</b><br>
<br>
En el ejemplo anterior hemos visto como <code>*args</code> puede ser iterado, ya que en realidad es una tupla. <br>
Por lo tanto iterando la tupla podemos acceder a todos los argumentos de entrada

<h3><code>**kwargs</code></h3>

Al igual que en <code>*args</code>, en <code>\**kwargs</code> el nombre es una mera convención entre los usuarios de Python.<br>
Puedes usar cualquier otro nombre siempre y cuando respetes el <b>**</b>.

````python
def imprimir_todos_parametro(**kwargs):
    for clave, valor in kwargs.items():
        print(clave, "=", valor)

imprimir_todos_parametro(Nombre='Tiziano', Apellido='Borrelli', Nacionalidad = 'Italiana')
````
